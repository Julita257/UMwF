```python
import pandas as pd
from sklearn.impute import KNNImputer

file_path = '/content/case1.csv'
df = pd.read_csv(file_path, delimiter=';')

print(df)
```

```python
df.head()
```

```python
df.drop('Address', axis=1, inplace=True)
df.drop('Unnamed: 0', axis=1, inplace=True)
df.drop('Index', axis=1, inplace=True)
df.head()
```

```python
# Sprawdzanie, które kolumny mają brakujące wartości
kolumny_z_brakami = df.isnull().any()

# Wyświetlenie kolumn z brakującymi wartościami
print(kolumny_z_brakami)
```

```python
# Definicja zakresu kolumn do uzupełnienia
kolumny_do_uzupelnienia = df.loc[:, 'Total ERC20 tnxs':'ERC20 uniq rec token name']

# Inicjalizacja KNNImputer
imputer = KNNImputer(n_neighbors=3)

# Uzupełnienie brakujących wartości w wybranych kolumnach
kolumny_uzupelnione = imputer.fit_transform(kolumny_do_uzupelnienia)

# Aktualizacja oryginalnej ramki danych uzupełnionymi danymi
df.loc[:, 'Total ERC20 tnxs':'ERC20 uniq rec token name'] = kolumny_uzupelnione
```

```python
# Lista kolumn tekstowych, dla których chcesz uzupełnić braki
kolumny_tekstowe = ['ERC20 most sent token type', 'ERC20_most_rec_token_type']

for kolumna in kolumny_tekstowe:
    # Znalezienie najczęściej występującej wartości w kolumnie
    moda = df[kolumna].mode()[0]

    # Uzupełnienie braków najczęściej występującym słowem
    df[kolumna] = df[kolumna].fillna(moda)
```

```python
# Sprawdzanie, które kolumny mają brakujące wartości
kolumny_z_brakami = df.isnull().any()

# Wyświetlenie kolumn z brakującymi wartościami
print(kolumny_z_brakami)
```

```python
# Usunięcie zduplikowanych wierszy, pozostawiając pierwsze wystąpienie
df = df.drop_duplicates()
```

```python
from sklearn.preprocessing import LabelEncoder
cols = ['ERC20 most sent token type', 'ERC20_most_rec_token_type']
# Encode labels of multiple columns at once
df[cols] = df[cols].apply(LabelEncoder().fit_transform)
# Print head
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

kolumny_do_skalowania = df.loc[:, 'Avg min between sent tnx':'ERC20 uniq rec token name'].columns

# Skalowanie wybranych kolumn
df[kolumny_do_skalowania] = scaler.fit_transform(df[kolumny_do_skalowania])
```
